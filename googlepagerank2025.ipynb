```python
import numpy as np


a = np.array([[0, 0.75, 0.25],



              [0.9, 0, 0.1],



              [0.7, 0.3, 0]])



mod = np.array([[1/3, 1/3, 1/3],



              [1/3, 1/3, 1/3],



              [1/3, 1/3, 1/3]])



b = np.array([[0.5,0.25,0.25]])

alpha = 0.5

kaisu = 10

akaisu = np.linalg.matrix_power(alpha*a+(1-alpha)*mod, kaisu)

c1=np.dot(b, akaisu)


print(c1)
```
