```
import networkx as nx

g = nx.DiGraph()

g.add_node(1)
g.add_node(2)
g.add_node(3)                                                                                                                     
g.add_node(4)
g.add_node(5)

g.add_node(6)
g.add_node(7)
g.add_node(8)


g.add_edge(1,2)
g.add_edge(2,1)
g.add_edge(2,3)
g.add_edge(3,1)
g.add_edge(3,4)
g.add_edge(3,5)
g.add_edge(4,1)
g.add_edge(4,5)
g.add_edge(5,1)
g.add_edge(5,2)
g.add_edge(5,4)

g.add_edge(4,6)
g.add_edge(4,7)
g.add_edge(4,8)
g.add_edge(6,4)
g.add_edge(6,7)
g.add_edge(6,8)
g.add_edge(7,4)
g.add_edge(7,6)
g.add_edge(7,8)
g.add_edge(8,4)
g.add_edge(8,6)
g.add_edge(8,7)


nx.draw_networkx(g)

pr=nx.pagerank(g,alpha=0.85)

print(pr)
```

```
#Listing sites that are possibly span sites

g1 = g.to_undirected()

nx.draw_networkx(g1)

Cliques = []
for c in nx.find_cliques(g1):
   Cliques.append( set(c) )

Cliques
```

```
import networkx as nx

g = nx.DiGraph()

g.add_node(1)
g.add_node(2)
g.add_node(3)                                                                                                                     
g.add_node(4)
g.add_node(5)

g.add_node(6)
g.add_node(7)
g.add_node(8)


g.add_edge(1,2)
g.add_edge(2,1)
g.add_edge(2,3)
g.add_edge(3,1)
g.add_edge(3,4)
g.add_edge(3,5)
g.add_edge(4,1)
g.add_edge(4,5)
g.add_edge(5,1)
g.add_edge(5,2)
g.add_edge(5,4)

g.add_edge(4,6)
g.add_edge(4,7)
g.add_edge(4,8)
g.add_edge(6,4)
g.add_edge(6,7)
g.add_edge(6,8)
g.add_edge(7,4)
g.add_edge(7,6)
g.add_edge(7,8)
g.add_edge(8,4)
g.add_edge(8,6)
g.add_edge(8,7)




nx.draw_shell(g)

pr=nx.pagerank(g,alpha=0.99)

print("-----pagerank-----")
print(pr)
```
