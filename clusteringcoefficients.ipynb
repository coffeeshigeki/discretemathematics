```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.complete_graph(5)

nx.draw_circular(G, node_size=10, node_color='red')

print(nx.clustering(G))

print(nx.average_clustering(G))
```

```
cg = nx.cycle_graph(18)
nx.draw_circular(cg, node_size=10, node_color='red')

print(nx.clustering(cg))

print(nx.average_clustering(cg))
```

```
ws = nx.watts_strogatz_graph(50, 10, 0.7)
nx.draw_circular(ws, node_size=10, node_color='red')

print(nx.clustering(ws))

print(nx.average_clustering(ws))
```

```
ba = nx.barabasi_albert_graph(100, 10)
nx.draw_circular(ba, node_size=10, node_color='red')

print(nx.clustering(ba))

print(nx.average_clustering(ba))
```
