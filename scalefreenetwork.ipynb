```
import networkx as nx
import matplotlib.pyplot as plt

ba = nx.barabasi_albert_graph(1000, 3)
plt.subplot(3,1,1)
nx.draw(ba, node_size=10, node_color='red')
print(nx.info(ba))
plt.subplot(3,1,2)
plt.plot(nx.degree_histogram(ba))
plt.subplot(3,1,3)
plt.xscale("log")
plt.yscale("log")
plt.grid(which="both")
plt.plot(nx.degree_histogram(ba))

nx.degree_histogram(ba)

import numpy as np

print(nx.degree_histogram(ba))

len(np.array(nx.degree_histogram(ba)))

list_a=[]
for x in range(len(np.array(nx.degree_histogram(ba)))):
  list_a.append(x)
print(list_a)


np.vstack((list_a,np.array(nx.degree_histogram(ba))))


```

```
import pandas as pd
df = pd.DataFrame(
    list_a
)
df[1] = nx.degree_histogram(ba)

df.columns = ['degree','number']

df
```

```
databeforelog = df[df['number'] != 0]
print(databeforelog)
```

```
import numpy
import pandas
dataafterlog = np.log(databeforelog)
print(np.log(databeforelog))
import matplotlib.pyplot as plt
dataafterlog.plot()
dataafterlog.plot.scatter(x='degree', y='number')
```

```
from sklearn.linear_model import LinearRegression

mod = LinearRegression(fit_intercept = True)
lm = mod.fit(dataafterlog[["degree"]], dataafterlog[["number"]])
print(lm)
 
print(mod.coef_)
print(mod.intercept_)

slope =  mod.coef_[0,0]
print(slope)

interception = mod.intercept_[0]

print(interception)



```

```
x = np.linspace( 0, 5, 10)   
y = slope * x + interception
dataafterlog.plot.scatter(x='degree', y='number',color="b")
plt.plot(x, y,color="r")
plt.legend()
```
